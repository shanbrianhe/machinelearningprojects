# MNIST Digit Classification with KNN and Naive Bayes

__What is MNIST?__

MNIST is a subset of the NIST, a database of handwritten digits that were collected from American Census Bureau employees and high school students. The MNIST database has 60,000 training and 10,000 testing images of handwritten digits which have been normalized by its creators to 28*28 pixel bounding boxes. 

MNIST is one of the most well known datasets for Machine Learning enthusiasts to work on.  It requires little data preprocessing and allows people to dive right in to practice machine-learning techniques in the context of image recognition. 

If you're interested, check out these links related to digit recognition:

Yann Lecun's MNIST benchmarks: http://yann.lecun.com/exdb/mnist/

Stanford Streetview research and data: http://ufldl.stanford.edu/housenumbers/

(2) Evaluate a K-Nearest-Neighbors model with k = [1,3,5,7,9] using the mini training set. Report accuracy on the dev set. For k=1, show precision, recall, and F1 for each label. Which is the most difficult digit?

- KNeighborsClassifier() for fitting and predicting
- classification_report() for producing precision, recall, F1 results

```python
def P2(k_values):

    ### STUDENT START ###
    
    for k_value in k_values:
        
        #initiate model instance
        KNN = KNeighborsClassifier(n_neighbors=k_value) 
        #fit training data & labels
        KNN.fit(mini_train_data, mini_train_labels) 
        #predict labels for dev data
        KNN_pred = KNN.predict(dev_data) 
        
        if k_value == 1: 
            print ("k_value =", k_value)
            print (classification_report(dev_labels, KNN_pred))
            print ("Accuracy =", (dev_labels == KNN_pred).mean())
        else:
            print ("k_value =", k_value)
            print ("Accuracy =", (dev_labels == KNN_pred).mean())
    
    ### STUDENT END ###

k_values = [1, 3, 5, 7, 9]

P2(k_values)
```

ANSWER: 9.0, with the lowest F1 score, seems to be the most difficult digit

(6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
import math

def P6():

    ### STUDENT START ###
    
    def BlurData(train_data):

        train_data_blur = np.zeros((len(train_data), 28, 28))
        train_data_reshape = train_data.reshape((len(train_data),28,28))

        #blur images with a Gaussian function of mean = 0 and var = 1
        d0 = 1/math.sqrt((2*math.pi))                   #weight for self
        d1 = 1/math.sqrt((2*math.pi)) *  math.exp(-0.5) #weight for adjacent cells
        d2 = 1/math.sqrt((2*math.pi)) *  math.exp(-2)   #weight for diagonally connected cells

        for x in range(0, len(train_data)):
            for i in range(1, 27):
                for j in range(1, 27):
                    #for each train_data[i][j], apply gaussian smoothing using the values of the 8 neighbouring pixels
                    train_data_blur[x][i][j] =  (d0 * train_data_reshape[x][i][j]) + 
                                                (d1 * (train_data_reshape[x][i-1][j] + 
                                                       train_data_reshape[x][i+1][j] + 
                                                       train_data_reshape[x][i][j-1] + 
                                                       train_data_reshape[x][i][j+1])) + 
                                                (d2 * (train_data_reshape[x][i-1][j+1] + 
                                                       train_data_reshape[x][i+1][j+1] + 
                                                       train_data_reshape[x][i-1][j-1] + 
                                                       train_data_reshape[x][i-1][j+1]))                
        
        return train_data_blur
    
    #preprocessed training data + original dev data
    KNN = KNeighborsClassifier(1)
    KNN.fit(BlurData(train_data).reshape((len(BlurData(train_data)), 784)), train_labels)
    KNN_pred = KNN.predict(dev_data)
    print ("Accuracy for preprocessed training data but not dev data:", (dev_labels == KNN_pred).mean())
    
    #orginal training data + preprocessed dev data
    KNN.fit(train_data, train_labels)
    KNN_pred = KNN.predict(BlurData(dev_data).reshape((len(dev_data), 784)))
    print ("Accuracy for preprocessed dev data but not training data:", (dev_labels == KNN_pred).mean())
    
    #preprocessed training data + preprocessed dev data
    KNN.fit(BlurData(train_data).reshape((len(BlurData(train_data)), 784)), train_labels)
    KNN_pred = KNN.predict(BlurData(dev_data).reshape((len(dev_data), 784)))
    print ("Accuracy for preprocessed training data and dev data:", (dev_labels == KNN_pred).mean())
    
    ## visualize blurring effect ## 
    print ("Example of the blurring effect:")
    
    mini_train_data_blur = BlurData(mini_train_data[:10]) # apply blurring function
    plt.subplot(1,2,1)
    plt.imshow(mini_train_data[6].reshape((28, 28))) # select a random original
    plt.title("Original")
    plt.subplot(1,2,2)
    plt.imshow(mini_train_data_blur[6]) # select the matching preprocessed digit
    plt.title("Blurred")
    
    ### STUDENT END ###
P6()
```

ANSWER: Comparing to having no preprocessing at all, having preprocessed training data but not dev data lost some accuracy but having dev data preprocessed made the model more accurate. More specifically, having both training data and dev data preprocessed achieved the highest accuracy. Gaussian smoothing improves the prediction accuracy by increasing the bias to help generalize better to dev data.

(9) Try training a model using GuassianNB, which is intended for real-valued features, and evaluate on the dev data. You'll notice that it doesn't work so well. Try to diagnose the problem. You should be able to find a simple fix that returns the accuracy to around the same rate as BernoulliNB. Explain your solution.

Hint: examine the parameters estimated by the fit() method, theta\_ and sigma\_.

```python
def P9():
### STUDENT END ###
    #original GaussianNB model
    GNB_1 = GaussianNB()
    GNB_1.fit(mini_train_data, mini_train_labels)
    GNB_1_PRED = GNB_1.predict(dev_data)
    
    #improved GaussianNB model
    GNB_2 = GaussianNB()
    GNB_2.fit(mini_train_data, mini_train_labels)
    GNB_2.sigma_ += .01
    GNB_2_PRED = GNB_2.predict(dev_data)
    
    print ('Accuracy of using a Gaussian Naive Bayes is:', ((dev_labels == GNB_1_PRED).mean()))
    print ('Accuracy of using an improved Gaussian Naive Bayes is:', ((dev_labels == GNB_2_PRED).mean()))
### STUDENT END ###
P9()
```

ANSWER: The original Gaussian Naive Bayes did poorly and it's likely due overfitting (with small sigma\_ values calculated by the Gaussian NB by default). Hence I added a constant to the sigma\_ value to increase the variance of the estimated Gaussian distribution to improve the generalization of the model. This simple fix improved the prediction accuracy significantly.

(10) Because Naive Bayes is a generative model, we can use the trained model to generate digits. Train a BernoulliNB model and then generate a 10x20 grid with 20 examples of each digit. Because you're using a Bernoulli model, each pixel output will be either 0 or 1. How do the generated digits compare to the training digits?

- You can use np.random.rand() to generate random numbers from a uniform distribution
- The estimated probability of each pixel is stored in feature\_log\_prob\_. You'll need to use np.exp() to convert a log probability back to a probability.

```python
def P10(num_examples):

### STUDENT START ###
    BNB = BernoulliNB()
    BNB.fit(mini_train_data, mini_train_labels)
    
    for digit in range(10):
        for num in range(1, num_examples+1):
            
            #contain non-processed predicted feature_log_prob_
            pixels = np.exp(BNB.feature_log_prob_)[digit]
            
            #use random number generator to set random pixels to 0
            pixels_bool = pixels >= np.random.random_sample((784,)) * 0.9
            pixels[pixels_bool == False] = 0  
            
            #use random number generator to add noise
            pixels = (pixels + np.random.rand()).reshape((28, 28))
            
            #plot genreated numbers
            plt.subplot(10, num_examples, digit * num_examples + num)
            plt.imshow(pixels)
            plt.axis('off')

### STUDENT END ###

P10(20)
```

ANSWER: After both 1) setting random pixels to 0 and 2) adding noise to the digits, the generated numbers look similar are less legible compared to most training samples. 

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    
    ### STUDENT START ###
    BNB = BernoulliNB(alpha = 0.1)
    BNB.fit(mini_train_data, mini_train_labels)
    BNB_pred = BNB.predict(dev_data)
    BNB_pred_prob = BNB.predict_proba(dev_data)
    
    for x in range(len(dev_data)):
        #caluate max posterior#
        max_post = BNB_pred_prob[x].max()
        
        #fill in correct count and total count
        for i in range(len(buckets)):
            if (i == 0 and max_post <= buckets[i]) | ( buckets[i-1] < max_post <= buckets[i]):
#             if max_post <= buckets[i]:
                total[i] += 1
                if BNB_pred[x] == dev_labels[x]:
                    correct[i] += 1
            else:
                continue
    
    return buckets, correct, total
            
    ### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
   accuracy = 0.0
   if (total[i] > 0): accuracy = correct[i] / total[i]
   print ('p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy))
```

ANSWER: This Naive Bayes model is weakly calibrated as the actual accuracy shows postive correlation with the posterior probability but they don't match each other at different levels of posterior probabilities.  

(12) EXTRA CREDIT

Try designing extra features to see if you can improve the performance of Naive Bayes on the dev set. Here are a few ideas to get you started:
- Try summing the pixel values in each row and each column.
- Try counting the number of enclosed regions; 8 usually has 2 enclosed regions, 9 usually has 1, and 7 usually has 0.

Make sure you comment your code well!

```python
def P12():

### STUDENT START ###
    #def preprocess function
    def preprocess (data):
        new_data = []
        for i in data:
            #insert row mean
            i_row_sum = np.append(i.reshape((28, 28)), 
                                  i.reshape((28, 28)).mean(axis = 1).reshape((28, 1)), 
                                  axis = 1)
            #insert column mean
            i_new = np.append(i_row_sum, 
                              i_row_sum.mean(axis = 0).reshape((1, 29)),
                              axis = 0)
            new_data.append(i_new.flatten())
        return new_data 
    
    BNB = BernoulliNB()
    BNB.fit(preprocess(mini_train_data), mini_train_labels)
    BNB_pred = BNB.predict(preprocess(dev_data))
    print ('Accuracy of using a Bernoulli Naive Bayes is:', ((dev_labels == BNB_pred).mean()))
    
### STUDENT END ###

P12()
```

I added the column and row means as extra features and it improved the accuracy by 1%. I chose Bernoulli NB since it outperformed the other NB's due to the natural distribution of the pixel values. But Bernoulli NB also binarizes the extra features, making the column and row sum essentially features that is 0 if there is only 0 values in the column/row or 1 if there is any non-zero value in the column/row.
